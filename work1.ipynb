```python
import seaborn as sns
import matplotlib.pyplot as plt

app_usage = df.groupby('App')['Usage (minutes)'].sum()

plt.figure(figsize=(10, 6))
plt.bar(app_usage.index, app_usage.values)
plt.xlabel('App')
plt.ylabel('Total Usage (minutes)')
plt.title('Total Usage per App')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
app_notifications = df.groupby('App')['Notifications'].sum()

plt.figure(figsize=(10, 6))
plt.bar(app_notifications.index, app_notifications.values)
plt.xlabel('App')
plt.ylabel('Total Notifications')
plt.title('Total Notifications per App')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
app_openings = df.groupby('App')['Times Opened'].sum()

plt.figure(figsize=(10, 6))
sns.barplot(x=app_openings.index, y=app_openings.values)
plt.xlabel('App')
plt.ylabel('Total Times Opened')
plt.title('Frequency of App Openings')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(data=df, x="Notifications", y="Usage (minutes)", hue="App", alpha=0.6)
plt.title("App Usage vs. Notifications")
plt.xlabel("Number of Notifications")
plt.ylabel("Usage (minutes)")
plt.legend(title='App')
plt.show()
```

```python
corr_matrix = df[['Usage (minutes)', 'Notifications', 'Times Opened']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation between Usage, Notifications, and Times Opened")
plt.show()
```

```python
sns.boxplot(data=df, x="App", y="Usage (minutes)")
plt.xticks(rotation=45)
plt.title("Usage Distribution per App")
plt.ylabel("Usage (minutes)")
plt.xlabel("App")
plt.show()
```

```python
app_opened = df.groupby("App")["Times Opened"].sum().sort_values()
app_opened.plot(kind='barh', color='skyblue', figsize=(10, 6))
plt.title("Total Times Opened per App")
plt.xlabel("Total Times Opened")
plt.ylabel("App")
plt.show()
```

```python
sns.scatterplot(data=df, x="Notifications", y="Times Opened", hue="App", alpha=0.6)
plt.title("Notifications vs Times Opened by App")
plt.xlabel("Notifications")
plt.ylabel("Times Opened")
plt.legend(title='App')
plt.show()
```
